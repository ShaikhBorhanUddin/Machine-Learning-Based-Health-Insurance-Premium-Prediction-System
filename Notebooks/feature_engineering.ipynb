```python
import numpy as np

# Define the updated conditions for each blood pressure category, with 'Low Blood Pressure' first
conditions = [
    (df_cleaned['systolic_bp'] <= 90) | (df_cleaned['diastolic_bp'] <= 60),
    (df_cleaned['systolic_bp'] < 120) & (df_cleaned['diastolic_bp'] < 80),
    (df_cleaned['systolic_bp'] >= 120) & (df_cleaned['systolic_bp'] <= 129) & (df_cleaned['diastolic_bp'] < 80),
    ((df_cleaned['systolic_bp'] >= 130) & (df_cleaned['systolic_bp'] <= 139)) | ((df_cleaned['diastolic_bp'] >= 80) & (df_cleaned['diastolic_bp'] <= 89)),
    (df_cleaned['systolic_bp'] >= 140) | (df_cleaned['diastolic_bp'] >= 90)
]

# Define the corresponding choices for each category in the correct order
choices = ['Low Blood Pressure', 'Normal', 'Elevated', 'Hypertension Stage 1', 'Hypertension Stage 2']

# Update the 'bp_category' column using np.select
df_cleaned['bp_category'] = np.select(conditions, choices, default='Unknown')
```

```python
import numpy as np

# Define bin edges and labels for 'bmi'
bmi_bins = [0, 18.5, 25, 30, 35, 40, float('inf')]
bmi_labels = ['Underweight', 'Normal', 'Overweight', 'Obese I', 'Obese II', 'Obese III']
df_cleaned['bmi_group'] = pd.cut(df_cleaned['bmi'], bins=bmi_bins, labels=bmi_labels, right=False)
```

```python
import numpy as np

# Define bin edges and labels for 'ldl'
ldl_bins = [0, 100, 130, 160, 190, np.inf]
ldl_labels = ['Optimal', 'Near Optimal', 'Borderline High', 'High', 'Very High']
df_cleaned['ldl_group'] = pd.cut(df_cleaned['ldl'], bins=ldl_bins, labels=ldl_labels, right=False)

# Define bin edges and labels for 'hba1c'
hba1c_bins = [0, 5.7, 6.5, np.inf]
hba1c_labels = ['Normal', 'Prediabetes', 'Diabetes']
df_cleaned['hba1c_group'] = pd.cut(df_cleaned['hba1c'], bins=hba1c_bins, labels=hba1c_labels, right=False)
```

```python
import numpy as np

# Define bin edges and labels for 'annual_premium'
annual_premium_bins = [0, 500, 2000, 5000, 10000, np.inf]
annual_premium_labels = ['Very Low', 'Low', 'Medium', 'High', 'Very High']
df_cleaned['annual_medical_cost_grouped'] = pd.cut(df_cleaned['annual_premium'], bins=annual_premium_bins, labels=annual_premium_labels, right=False)
```
