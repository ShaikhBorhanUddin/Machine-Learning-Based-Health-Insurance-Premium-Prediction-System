```python
columns_to_drop = ['person_id', 'provider_quality', 'risk_score', 'monthly_premium', 'policy_changes_last_2yrs', 'claims_count', 'avg_claim_amount', 'total_claims_paid', 'chronic_count', 'proc_imaging_count', 'proc_surgery_count', 'proc_physio_count', 'proc_consult_count', 'proc_lab_count', 'is_high_risk']

df_cleaned = df.drop(columns=columns_to_drop)
```

```python
df_cleaned['alcohol_freq'] = df_cleaned['alcohol_freq'].fillna('Never')
```

```python
for col in df_cleaned.select_dtypes(include=['object']).columns:
    print(f"Unique values in '{col}': {df_cleaned[col].unique()}")
```

```python
df_cleaned['education'] = df_cleaned['education'].replace({'No HS': 'High School Dropout', 'HS': 'High School', 'Some College': 'College'})
```

```python
df_cleaned['plan_type'] = df_cleaned['plan_type'].replace({'PPO': 'Preferred Provider Organization', 'POS': 'Point-of-Service', 'HMO': 'Health Maintenance Organization', 'EPO': 'Exclusive Provider Organization'})
```

```python
for col in df_cleaned.select_dtypes(include=['object']).columns:
    print(f"Unique values in '{col}': {df_cleaned[col].unique()}")
```
